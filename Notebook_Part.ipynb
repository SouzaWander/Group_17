### Method 3: Plots an area chart of the consumption

The method present below plots an area chart of the consumption, as argument have two arguments the first one is the name of the coutry name the second one works as option if its true normalizes the consumption values otherwise it plots the absolute value.

```python
#Consumption unormalized
object.show_consumption("Portugal", False)
```

```python
#Consuption normalized
object.show_consumption("Portugal", True)
```

### Method 4: Plot the total consumption of the selected countries

The fourth method receive a list of country strings. After the method compare the total of the "_consumption" columns for each of the chosen countries and plot it, so a comparison can be made.

```python
object.consumption_country(["Switzerland", "Portugal", "Chile"])
```

### Method 5: Plot the gdp of the selected countries over time

The fifth method receive a list of country strings. After the method compare the "gdp" column of each country over the years.

```python
object.gdp_country(["Switzerland", "Portugal", "Chile"])
```

### Method 6: Plot a Scatter Plot comparing the Total Energy Consumption, GDP and the Population of each Country

In this method, we asked the user to insert a year on the method we called 'gapminder'. The user could also compare the population of each country by the size of data point. 


```python
object.gapminder(2010)
```
